### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", delimiter="|")
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation")["age"].agg(["mean"]).sort_values(by="mean", ascending=False)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users.head()
genders = users.groupby(["occupation","gender"], as_index=False)[["user_id"]].agg("count")
occ_genders = users.groupby("occupation", as_index=False)[["user_id"]].agg("count")
genders = genders.merge(occ_genders, how="left", left_on="occupation", right_on="occupation")
genders["male_ratio"] = genders.user_id_x/genders.user_id_y
genders[genders["gender"]=="M"].sort_values(by="male_ratio", ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.head()
users.groupby("occupation")["age"].agg(["min","max"])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation","gender"])["age"].agg(["mean","median"]).sort_values(by="mean", ascending=False)
```

### Step 8.  For each occupation present the percentage of women and men

```python
users.head()
genders = users.groupby(["occupation","gender"], as_index=False)[["user_id"]].agg("count")
occ_genders = users.groupby("occupation", as_index=False)[["user_id"]].agg("count")
genders = genders.merge(occ_genders, how="left", left_on="occupation", right_on="occupation")
genders["percentage"] = (genders.user_id_x/genders.user_id_y)*100
genders
```
